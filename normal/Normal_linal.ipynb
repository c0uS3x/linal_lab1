```python
import itertools

def subgroups_of_Sm(ISU: int) -> dict:
    N = ISU % 20
    m = 4 + (N % 5)
    
    elements = list(itertools.permutations(range(1, m + 1)))
    identity = tuple(range(1, m + 1))
    
    def multiply(p1, p2):
        return tuple(p1[p2[i]-1] for i in range(len(p1)))
    
    def inverse(p):
        inv = [0] * len(p)
        for i, val in enumerate(p):
            inv[val-1] = i + 1
        return tuple(inv)
    
    known_counts = {3: 6, 4: 30, 5: 156, 6: 1455, 7: 11300}
    total_subgroups_count = known_counts.get(m, "")
    
    cyclic_subgroups = []
    seen_subgroups = set()
    
    cyclic_subgroups.append({identity})
    seen_subgroups.add(frozenset({identity}))
    
    sorted_elements = sorted(elements)
    
    for g in sorted_elements:
        if g == identity:
            continue
        
        current_subgroup = set()
        curr = g
        while curr not in current_subgroup:
            current_subgroup.add(curr)
            curr = multiply(curr, g)
            
        fs = frozenset(current_subgroup)
        if fs not in seen_subgroups:
            seen_subgroups.add(fs)
            cyclic_subgroups.append(current_subgroup)
            
        if isinstance(total_subgroups_count, int) and len(cyclic_subgroups) > (N % total_subgroups_count) + 50:
            break

    if isinstance(total_subgroups_count, int):
        target_idx = N % total_subgroups_count
    else:
        target_idx = N  
        
    target_subgroup = list(cyclic_subgroups[target_idx])
    
    H = set(target_subgroup)
    H_list = sorted(list(H))
    
    group_order = len(elements)
    subgroup_order = len(H)
    subgroup_index = group_order // subgroup_order
    
    left_cosets = []
    right_cosets = []

    unique_left = set()
    unique_right = set()
    
    for g in elements:
        l_coset = frozenset(multiply(g, h) for h in H_list)
        if l_coset not in unique_left:
            unique_left.add(l_coset)
            left_cosets.append(sorted(list(l_coset)))
            
        r_coset = frozenset(multiply(h, g) for h in H_list)
        if r_coset not in unique_right:
            unique_right.add(r_coset)
            right_cosets.append(sorted(list(r_coset)))

    is_normal = (unique_left == unique_right)

    return {
        "N": N,
        "m": m,
        "group": f"S_{m}",
        "total_subgroups_count": total_subgroups_count,
        "target_subgroup_index": target_idx,
        "target_subgroup_elements": H_list,
        "subgroup_order": subgroup_order,
        "subgroup_index": subgroup_index,
        "is_normal": is_normal,
        "left_cosets_count": len(left_cosets),
        "right_cosets_count": len(right_cosets),
        "sample_left_coset": left_cosets[1] if len(left_cosets) > 1 else left_cosets[0],
        "sample_right_coset": right_cosets[1] if len(right_cosets) > 1 else right_cosets[0]
    }

result = subgroups_of_Sm(504978)
import pprint
pprint.pprint(result)
```

```python
import math
import itertools

def element_powers_in_Sm(N: int) -> dict:
    N_mod_20 = N % 20
    m = 4 + (N_mod_20 % 5)
    
    n1 = N_mod_20 % 6
    n2 = (N_mod_20 + 1) % 6
    n3 = (N_mod_20 + 2) % 6
    
    all_permutations = list(itertools.permutations(range(1, m + 1)))
    group_size = len(all_permutations)

    target_index = N % group_size
    g = all_permutations[target_index]
    
    identity = tuple(range(1, m + 1))
    
    def multiply(p1, p2):
        return tuple(p1[p2[i]-1] for i in range(len(p1)))
    
    def power(p, k):
        if k == 0:
            return identity
        res = identity
        base = p
        for _ in range(k):
            res = multiply(res, base)
        return res

    def get_order(p):
        if p == identity:
            return 1
        curr = p
        k = 1
        while curr != identity:
            curr = multiply(curr, p)
            k += 1
        return k

    elem1 = power(g, n1)
    elem2 = power(g, n2)
    elem3 = power(g, n3)
    
    ord1 = get_order(elem1)
    ord2 = get_order(elem2)
    ord3 = get_order(elem3)

    return {
        "N (ISU)": N,
        "m": m,
        "Group Order |Sm|": group_size,
        "Target Index (g)": target_index,
        "Element g": g,
        "Powers (n1, n2, n3)": (n1, n2, n3),
        "Elements": {
            f"g^{n1}": elem1,
            f"g^{n2}": elem2,
            f"g^{n3}": elem3
        },
        "Orders": {
            f"order(g^{n1})": ord1,
            f"order(g^{n2})": ord2,
            f"order(g^{n3})": ord3
        },
        "Subgroup Orders": {
            f"|<g^{n1}>|": ord1,
            f"|<g^{n2}>|": ord2,
            f"|<g^{n3}>|": ord3
        }
    }


result = element_powers_in_Sm(504978)
pprint.pprint(result)
```

```python
import math

def cyclic_subgroup_in_Zm_additive(ISU: int) -> dict:
    N_mod_20 = ISU % 20
    m = 4 + (N_mod_20 % 5)
    
    t = 12
    
    element = t % m
    
    subgroup = set()
    curr = 0
    while True:
        subgroup.add(curr)
        next_val = (curr + element) % m
        if next_val in subgroup:
            break
        curr = next_val
        
    sorted_subgroup = sorted(list(subgroup))
    order = len(sorted_subgroup)
    
    generators = []
    for x in sorted_subgroup:
        temp_set = set()
        temp_val = 0
        while True:
            temp_set.add(temp_val)
            nxt = (temp_val + x) % m
            if nxt in temp_set:
                break
            temp_val = nxt
        
        if len(temp_set) == order:
            generators.append(x)

    calc_order = m // math.gcd(t, m)
    
    return {
        "m": m,
        "t_original": t,
        "element (t mod m)": element,
        "subgroup_elements": sorted_subgroup,
        "subgroup_order": order,
        "generators_of_subgroup": generators,
        "relation_check": f"Order calculated by formula m/gcd(t,m) is {calc_order}"
    }

import pprint
pprint.pprint(cyclic_subgroup_in_Zm_additive(504978))
```

```python
from sympy import symbols, Poly, GF

ISU = 504978
N = ISU % 20 

print(f"Корни полиномов (N={N})\n")

# Поле F4 

def add_f4(a, b):
    return a ^ b

MUL_TABLE = [
    [0, 0, 0, 0],
    [0, 1, 2, 3],
    [0, 2, 3, 1],
    [0, 3, 1, 2]
]

def mul_f4(a, b):
    return MUL_TABLE[a][b]

def pow_f4(base, exp):
    res = 1
    for _ in range(exp):
        res = mul_f4(res, base)
    return res

coeffs = {} 
coeffs[9] = 1

print("Коэффициенты полинома:")
print(f"x^9: 1")
for i in range(9):
    val = (i + N) % 4
    coeffs[i] = val
    if val != 0:
        print(f"x^{i}: {val}")

def evaluate_poly(x_val):
    result = 0
    for power, coeff in coeffs.items():
        if coeff == 0: continue
        term_val = mul_f4(coeff, pow_f4(x_val, power))
        result = add_f4(result, term_val)
    return result

found_roots = []
elements = [0, 1, 2, 3]
names = {0: "0", 1: "1", 2: "2 (alpha)", 3: "3 (alpha+1)"}

for elem in elements:
    val = evaluate_poly(elem)
    if val == 0:
        found_roots.append(elem)

if found_roots:
    print("\nНАЙДЕННЫЕ КОРНИ:")
    for r in found_roots:
        print(f"-> Элемент {names[r]}")
        
    print("(Проверка кратности: если корень обнуляет и производную, кратность > 1)")
    deriv_coeffs = {}
    for p, c in coeffs.items():
        if p % 2 != 0:
            deriv_coeffs[p-1] = c
            
    def evaluate_deriv(x_val):
        res = 0
        for p, c in deriv_coeffs.items():
            term = mul_f4(c, pow_f4(x_val, p))
            res = add_f4(res, term)
        return res
        
    for r in found_roots:
        if evaluate_deriv(r) == 0:
            print(f"  Внимание: Корень {names[r]} может иметь кратность > 1")
        else:
            print(f"  Корень {names[r]} имеет кратность 1")

else:
    print("\nКорней в F4 нет.")


# Поле F7 
x = symbols('x')
coeffs_b = []
degree_b = 6
for j in range(degree_b + 1):
    val = (j + N) % 7
    coeffs_b.append(val)

coeffs_full_b = list(reversed(coeffs_b))

f7_poly = Poly(coeffs_full_b, x, domain=GF(7))
print(f"Полином f(x) над F7: {f7_poly.as_expr()}")

roots_7 = f7_poly.ground_roots()
if roots_7:
    print(f"Корни {{значение: кратность}}: {roots_7}")
else:
    print("  Корней нет.")
```

```python
from sympy import symbols, Poly, GF
import itertools

def generate_irreducible_polynomials(q: int, d: int) -> list:
    """
    Генерирует список всех неприводимых полиномов степени d над полем F_q.
    
    Args:
        q (int): Порядок простого поля (простое число).
        d (int): Степень полинома.
        
    Returns:
        list: Список полиномов (выражений sympy), которые являются неприводимыми.
    """
    x = symbols('x')
    domain = GF(q)
    irreducibles = []
    
    leading_coeffs = range(1, q)
    other_coeffs = range(q)
    
    for a_d in leading_coeffs:
        for remaining in itertools.product(other_coeffs, repeat=d):
            coeffs = (a_d,) + remaining
            
            f = Poly.from_list(coeffs, gens=x, domain=domain)
            
            if f.is_irreducible:
                irreducibles.append(f.as_expr())

    return irreducibles

if __name__ == "__main__":
    test_cases = [
        (2, 2), (2, 3), (2, 4),
        (3, 2), (3, 3), # (3, 4)
        (5, 2)          
    ]

    for q, d in test_cases:
        result = generate_irreducible_polynomials(q, d)
        count = len(result)
        
        print(f"Поле F_{q}, Степень d={d}")
        print(f"Всего найдено: {count}")
        
        if count <= 10:
            print("Список:", result)
        else:
            print("Первые 5:", result[:5], "...")
        print("-" * 60)

    large_tests = [(5, 3), (5, 4)]
    for q, d in large_tests:
        result = generate_irreducible_polynomials(q, d)
        print(f"Поле F_{q}, Степень d={d}")
        print(f"Всего найдено: {len(result)}")
        print("Пример:", result[0])
```
